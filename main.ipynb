```python
from sklearn.model_selection import train_test_split

X = diabetes.data
y = diabetes.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=10, test_size=0.2
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train)

print(lr.coef_)
print(lr.intercept_)

lr.predict(X_test)
```

```python
class batchGD:

    def __init__(self, learning_rate, epochs):
        self.coef_ = None
        self.intercept_ = None
        self.epochs = epochs
        self.learning_rate = learning_rate

    def fit(self, X_train, y_train):
        
        X_train = np.asarray(X_train, dtype=np.float64)
        y_train = np.asarray(y_train, dtype=np.float64).ravel()

        self.coef_ = np.zeros(X_train.shape[1])
        self.intercept_ = 0.0

        for i in range(self.epochs):
             
            y_hat = np.dot(X_train, self.coef_) + self.intercept_ 

            coef_loss = -2 * np.dot((y_train - y_hat), X_train) / X_train.shape[0]  
            intercept_loss = -2 * np.mean(y_train - y_hat) 

            self.coef_ -= self.learning_rate * coef_loss
            self.intercept_ -= self.learning_rate * intercept_loss
        
        return self
    
    def predict(self, X):
        X  = np.asarray(X, dtype=np.float64)
        return np.dot(X, self.coef_) + self.intercept_


        
```

```python
bgd = batchGD(0.01, 200000)

bgd.fit(X_train, y_train)
print(bgd.coef_)
print(bgd.intercept_)

bgd.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

my_pred = bgd.predict(X_test)
sklearn_pred = lr.predict(X_test)

print(mean_squared_error(y_test, my_pred))
print(mean_squared_error(y_test, sklearn_pred))
```

```python
print(np.max(np.abs(bgd.predict(X_train) - lr.predict(X_train))))
```

```python
from sklearn.metrics import r2_score

print(r2_score(y_test, bgd.predict(X_test)))
print(r2_score(y_test, lr.predict(X_test)))
```
