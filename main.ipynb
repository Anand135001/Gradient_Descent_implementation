```python
from sklearn.model_selection import train_test_split

X = diabetes.data
y = diabetes.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=10, test_size=0.2
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train)

print(lr.coef_)
print(lr.intercept_)
```

```python
class batchGD:

    def __init__(self, learning_rate, epochs):
        self.coef_ = None
        self.intercept_ = None
        self.epochs = epochs
        self.learning_rate = learning_rate

    def fit(self, X_train, y_train):
        
        self.coef_ = np.zeros(X_train.shape[1])
        self.intercept_ = 0 

        for i in range(self.epochs):
             
            y_hat = np.dot(X_train, self.coef_) + self.intercept_ 

            cofe_loss = -2 * np.dot((y_train - y_hat), X_train) / X_train.shape[0]  
            self.coef_ = self.coef_ - (self.learning_rate * cofe_loss)

            intercept_loss = -2 * np.mean(y_train - y_hat) 
            self.intercept_ = self.intercept_ - (self.learning_rate * intercept_loss)
        
        print(self.coef_, self.intercept_)


        
```

```python
bgd = batchGD(0.99, 18000)

bgd.fit(X_train, y_train)
```
