```python
from sklearn.datasets import load_diabetes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
diabetes = load_diabetes()
diabetes

df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = diabetes.data
y = diabetes.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=10, test_size=0.2
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train)

print(lr.coef_)
print(lr.intercept_)

lr.predict(X_test)
```

```python
class SGDRegressor:

    def __init__(self, learning_rate, epochs):
        self.coef_ = None
        self.intercept_ = None
        self.epochs = epochs
        self.learning_rate = learning_rate

    def fit(self, X_train, y_train):
        
        X_train = np.asarray(X_train, dtype=np.float64)
        y_train = np.asarray(y_train, dtype=np.float64).ravel()

        n_samples, n_features = X_train.shape

        self.coef_ = np.zeros(n_features)
        self.intercept_ = 0.0

        for i in range(self.epochs):
            for _ in range(n_samples):
                index = np.random.randint(0, n_samples)

                x_i = X_train[index]
                y_i = y_train[index]
                
                y_hat = np.dot(x_i, self.coef_ ) + self.intercept_

                intercept_der = -2 * (y_i- y_hat)
                coef_der = -2 * (y_i - y_hat) * x_i
 
                self.intercept_ -= self.learning_rate * intercept_der
                self.coef_ -= self.learning_rate * coef_der 

        return self
    
    def predict(self, X):
        X  = np.asarray(X, dtype=np.float64)
        return np.dot(X, self.coef_) + self.intercept_
```

```python
model = SGDRegressor(0.01, 90)

model.fit(X_train, y_train)

print(model.coef_)
print(model.intercept_)
```

```python
from sklearn.metrics import r2_score

print(r2_score(y_test, model.predict(X_test)))
print(r2_score(y_test, lr.predict(X_test)))
```
